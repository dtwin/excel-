```python
tables = []

def Read_Excel(excel):
    #
    for rows in range(3, excel.nrows-1):
        dict_ = {"日期":"", "区域":"", "记录人":"", "交接内容":"", "进度":"", "备注":"", "处理意见":"","类型":"","处理人":""}
        dict_["日期"] = table.cell_value(rows, 1)
        dict_["区域"] = table.cell_value(rows, 2)
        dict_["记录人"] = table.cell_value(rows, 3)
        dict_["交接内容"] = table.cell_value(rows, 4)
        dict_["进度"] = table.cell_value(rows, 5)
        dict_["备注"] = table.cell_value(rows, 6)
        dict_["处理意见"] = table.cell_value(rows, 7)
        dict_["类型"] = table.cell_value(rows, 8)
        dict_["处理人"] = table.cell_value(rows, 9)
        #
        tables.append(dict_)
```

```python
Read_Excel(table1)
for i in tables:
    print(i)
```
